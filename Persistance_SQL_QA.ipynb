```sql
WITH cte AS (
    SELECT 
        name,
        date,
        ROW_NUMBER() OVER (PARTITION BY name ORDER BY date) AS rn  -- We remove all 'N' rows and apply number 1,2,3 for each person
    FROM your_table
    WHERE visited_shop = 'Y'
),
grouped AS (
    SELECT 
        name,
        date,
        DATEADD(day, -rn, date) AS grp   -- use DATE_SUB in MySQL 👉 grp = date - rn
    FROM cte
)
SELECT name
FROM grouped
GROUP BY name, grp
HAVING COUNT(*) >= 4;
```

```sql
WITH cte AS (
    SELECT 
        train,
        start_stop,
        end_stop,
        kms,
        LAG(end_stop) OVER (PARTITION BY train ORDER BY start_stop) AS prev_end
    FROM your_table
)
SELECT 
    train,
    SUM(kms) AS total_kms
FROM cte
WHERE prev_end IS NULL 
   OR prev_end = start_stop
GROUP BY train;
```
